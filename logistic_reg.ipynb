# Logistic regression 

The labels needs to be binary (0 or 1 ) , since we use the loss of cross entropy 

```python
# my data
X = np.linspace(0,10,50)

print("Before: ", X.shape)

X_column = X.reshape(-1,1)  #2D Column vector
print("After: ", X_column.shape)
print(X_column[:3])

# -1 in reshape to find the compatibility with original shape
# -1 is the unkown dimension 
```

```python
# add intercept
ones = np.ones(len(X))
X_design = np.c_[ones,X_column]

# y_data
# Change the noise to be more accurate
noise = np.random.normal(0, 5, 50) 

# Equation
true_line = 2*X+1
y_continuous = 2*X +1 + noise
y = (y_continuous > (2*X + 1)).astype(int)
```

```python
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Plot class 0 and class 1 separately so we can label them
plt.scatter(X[y == 0], y_continuous[y == 0],
            marker='x', s=100, alpha=0.9, label='class 0 (below line)')
plt.scatter(X[y == 1], y_continuous[y == 1],
            marker='o', s=100, alpha=0.9, label='class 1 (above line)')

# True underlying line
plt.plot(X, true_line, color='grey', linewidth=2, label='True line (2x + 1)')

# Labels, legend, title
plt.xlabel("x")
plt.ylabel("continuous y")
plt.title("Continuous data colored by class (0/1)  better visualization for logistic")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
